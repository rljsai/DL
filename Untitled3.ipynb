```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import mne
import torch.nn.functional as F
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from torch.optim.lr_scheduler import ReduceLROnPlateau

# 1. Data Loading and Preprocessing
def load_eeg_data(data_dir, metadata_file):
    metadata = pd.read_csv(metadata_file)
    raw_data_rest = {}
    raw_data_test = {}
    
    for subject_id in range(1, 37):
        rest_file_path = os.path.join(data_dir, f'Subject{subject_id:02d}_1.edf')
        test_file_path = os.path.join(data_dir, f'Subject{subject_id:02d}_2.edf')
        
        if os.path.exists(rest_file_path) and os.path.exists(test_file_path):
            raw_rest = mne.io.read_raw_edf(rest_file_path, preload=True)
            raw_test = mne.io.read_raw_edf(test_file_path, preload=True)
            
            raw_data_rest[subject_id] = raw_rest
            raw_data_test[subject_id] = raw_test
    
    return raw_data_rest, raw_data_test, metadata
```

```python
def calculate_psd_welch(raw, bands):
    spectrum = raw.compute_psd(method='welch', fmin=1, fmax=100, n_fft=256, n_overlap=128)
    psd, freqs = spectrum.get_data(return_freqs=True)
    band_powers = {}
    for band, (fmin, fmax) in bands.items():
        idx_band = (freqs >= fmin) & (freqs <= fmax)
        band_power = psd[:, idx_band].mean(axis=1)
        band_powers[band] = band_power
    return band_powers
```

```python
bands = {
        "Delta": (1, 4),
        "Theta": (4, 8),
        "Alpha": (8, 13),
        "Beta": (13, 30),
        "Gamma": (30, 100)
    }
    
    # Calculate PSD
all_band_powers_rest = {subject_id: calculate_psd_welch(raw, bands) for subject_id, raw in raw_data_rest.items()}
all_band_powers_test = {subject_id: calculate_psd_welch(raw, bands) for subject_id, raw in raw_data_test.items()}
    
```

```python
def plot_band_powers(band_powers_rest, band_powers_test, bands):
    bands_list = list(bands.keys())
    band_indices = np.arange(len(bands_list))
    bar_width = 0.35
    
    fig, ax = plt.subplots(figsize=(10, 6))
    
    rests = [band_powers_rest[band].mean() for band in bands_list]
    tests = [band_powers_test[band].mean() for band in bands_list]
    
    ax.bar(band_indices - bar_width/2, rests, bar_width, label='Rest')
    ax.bar(band_indices + bar_width/2, tests, bar_width, label='Test')
    
    ax.set_xlabel('Frequency Bands')
    ax.set_ylabel('Mean Power')
    ax.set_title('Mean Power in Frequency Bands')
    ax.set_xticks(band_indices)
    ax.set_xticklabels(bands_list)
    ax.legend()
    
    plt.show()

```

```python
# 3. Feature Extraction
def prepare_data(all_band_powers):
    features = []
    for subject_id, band_powers in all_band_powers.items():
        subject_features = []
        for band, power in band_powers.items():
            subject_features.append(power.squeeze())
        features.append(np.array(subject_features))
    return torch.tensor(features, dtype=torch.float32)
```

```python
class EEGNet(nn.Module):
    def __init__(self, num_channels, num_samples, num_classes):
        super(EEGNet, self).__init__()
        self.conv1 = nn.Conv2d(1, 16, (num_channels, 1), padding=0, bias=False)
        self.batchnorm1 = nn.BatchNorm2d(16)
        self.conv2 = nn.Conv2d(16, 32, (1, 51), padding=(0, 25), bias=False)
        self.batchnorm2 = nn.BatchNorm2d(32)
        self.conv3 = nn.Conv2d(32, 32, (1, 15), padding=(0, 7), bias=False)
        self.batchnorm3 = nn.BatchNorm2d(32)
        self.pool = nn.AvgPool2d((1, 4))
        self.dropout = nn.Dropout(0.5)
        self.fc = nn.Linear(32 * (num_samples // 4), num_classes)

    def forward(self, x):
        x = x.unsqueeze(1)  # Add channel dimension
        x = torch.relu(self.batchnorm1(self.conv1(x)))
        x = torch.relu(self.batchnorm2(self.conv2(x)))
        x = torch.relu(self.batchnorm3(self.conv3(x)))
        x = self.pool(x)
        x = self.dropout(x)
        x = x.view(x.size(0), -1)
        x = self.fc(x)
        return x
```

```python
class TSCeption(nn.Module):
    def __init__(self, num_channels, num_samples, num_classes, sampling_rate=128):
        super(TSCeption, self).__init__()
        self.inception_window = [0.5, 0.25, 0.125]
        self.pool = 8
        self.sampling_rate = sampling_rate

        # Inception Modules
        self.inception_1 = nn.ModuleList()
        self.inception_2 = nn.ModuleList()
        self.inception_3 = nn.ModuleList()
        for i in range(len(self.inception_window)):
            kernel_size = int(self.inception_window[i] * self.sampling_rate)
            self.inception_1.append(nn.Conv2d(1, 16, (num_channels, kernel_size), padding=(0, kernel_size // 2)))
            self.inception_2.append(nn.Conv2d(16, 32, (1, kernel_size), padding=(0, kernel_size // 2)))
            self.inception_3.append(nn.Conv2d(32, 64, (1, kernel_size), padding=(0, kernel_size // 2)))

        # We'll set the size of fc1 later
        self.fc1 = None
        self.fc2 = nn.Linear(128, num_classes)

    def forward(self, x):
        
        x = x.unsqueeze(1)  # Add channel dimension
       
        # Inception Modules
        y = []
        for i in range(len(self.inception_window)):
            z = F.elu(self.inception_1[i](x))
            z = F.elu(self.inception_2[i](z))
            z = F.elu(self.inception_3[i](z))
            y.append(z)
        
        x = torch.cat(y, dim=1)
        
        x = F.avg_pool2d(x, (1, self.pool))
        
        x = x.view(x.size(0), -1)


        
        
        # Initialize fc1 if it's the first forward pass
        if self.fc1 is None:
            self.fc1 = nn.Linear(x.shape[1], 128)
        
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
def train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs, patience):
    best_val_loss = float('inf')
    trigger_times = 0
    for epoch in range(num_epochs):
        model.train()
        train_loss = 0.0
        for inputs, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
        
        model.eval()
        val_loss = 0.0
        val_preds = []
        val_labels = []
        with torch.no_grad():
            for inputs, labels in val_loader:
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                _, predicted = torch.max(outputs, 1)
                val_preds.extend(predicted.tolist())
                val_labels.extend(labels.tolist())
        
        with torch.no_grad():
            for inputs, labels in val_loader:
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                _, predicted = torch.max(outputs, 1)
                val_preds.extend(predicted.tolist())
                val_labels.extend(labels.tolist())
        train_loss /= len(train_loader)
        val_loss /= len(val_loader)
        scheduler.step(val_loss)
        
        print(f'Epoch [{epoch+1}/{num_epochs}], Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}')
        
        if val_loss < best_val_loss * 0.995:
            best_val_loss = val_loss
            trigger_times = 0
            torch.save(model.state_dict(), 'best_model.pth')
        else:
            trigger_times += 1
            if trigger_times >= patience:
                print("Early stopping!")
                break
    
    return val_preds, val_labels
        
```

```python
def evaluate_model(true_labels, predictions):
    accuracy = accuracy_score(true_labels, predictions)
    precision = precision_score(true_labels, predictions, average='weighted')
    recall = recall_score(true_labels, predictions, average='weighted')
    f1 = f1_score(true_labels, predictions, average='weighted')
    cm = confusion_matrix(true_labels, predictions)
    
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-score: {f1:.4f}")
    
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion Matrix')
    plt.colorbar()
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()
```

```python
X_rest = prepare_data(all_band_powers_rest)
X_test = prepare_data(all_band_powers_test)
y_rest = torch.zeros(X_rest.shape[0], dtype=torch.long)
y_test = torch.ones(X_test.shape[0], dtype=torch.long)
    
X = torch.cat([X_rest, X_test], dim=0)
y = torch.cat([y_rest, y_test], dim=0)
    
    # Set up k-fold cross-validation
k_folds = 5
kf = KFold(n_splits=k_folds, shuffle=True, random_state=42)
    
    # Model parameters
# Model parameters
num_channels = X.shape[1]  # This should be 5 (for the 5 frequency bands)
num_samples = X.shape[2]  # This should be 105 (or whatever the number of time points is)
num_classes = 2
    
    # Training parameters
num_epochs = 200
batch_size = 32
learning_rate = 0.0001
patience = 30
```

```python
for model_class in [EEGNet, TSCeption]:
        print(f"Training {model_class.__name__}")
        all_val_preds = []
        all_val_labels = []
        
        for fold, (train_idx, val_idx) in enumerate(kf.split(X)):
            print(f"Fold {fold + 1}")
            
            X_train, X_val = X[train_idx], X[val_idx]
            y_train, y_val = y[train_idx], y[val_idx]
            
            train_dataset = TensorDataset(X_train, y_train)
            val_dataset = TensorDataset(X_val, y_val)
            
            train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
            val_loader = DataLoader(val_dataset, batch_size=batch_size)
            
            model = model_class(num_channels, num_samples, num_classes)
            criterion = nn.CrossEntropyLoss()
            optimizer = optim.Adam(model.parameters(), lr=learning_rate)
            scheduler = ReduceLROnPlateau(optimizer, 'min', patience=5, factor=0.5)
            
            val_preds, val_labels = train_model(model, train_loader, val_loader, criterion, optimizer, scheduler, num_epochs, patience)
            
            all_val_preds.extend(val_preds)
            all_val_labels.extend(val_labels)
        
        print(f"Overall performance for {model_class.__name__}:")
        evaluate_model(all_val_labels, all_val_preds)
```
